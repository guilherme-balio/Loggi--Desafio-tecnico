```python
# Atribuindo o código de cada pacote à uma variável e depois em uma lista com todos os pacotes

cod1 = "888555555123888"
cod2 = "333333555584333"
cod3 = "222333555124000"
cod4 = "000111555874555"
cod5 = "111888555654777"
cod6 = "111333555123333"
cod7 = "555555555123888"
cod8 = "888333555584333"
cod9 = "111333555124000"
cod10 = "333888555584333"
cod11 = "555888555123000"
cod12 = "111888555123555"
cod13 = "888000555845333"
cod14 = "000111555874000"
cod15 = "111333555123555"

lista = [cod1,cod2,cod3,cod4,cod5,cod6,cod7,cod8,cod9,cod10,cod11,cod12,cod13,cod14,cod15]
```

```python
# importando biblioteca para separar os códigos de barra
from textwrap import wrap

# definindo função para "cortar" os códigos de barra

def corta(x):
    
    # criando lista vazia para guardar os códigos cortados
    cortados = []
    
    # loop levando em conta a quantidade de pacotes
    for n in range(0,len(x)):
        
        # usando função "wrap" para "cortar" o código a cada 3 caracteres
        y = wrap(x[n],3)
        
        # adicionando o código de barra "cortado" à lista "cortados"
        cortados.append(y)
    
    # retornando a lista completa
    return cortados

# aplicando função à lista de códigos feita e salvando em uma variável
cortado = corta(lista)
```

```python
# importando biblioteca pandas para se trabalhar com os dados
import pandas as pd

# criando um DataFrame com as os códigos de barra "cortados" e com a devida ordem de Trincas
df1 = pd.DataFrame(cortado, columns =["Região de Origem",
                                   "Região de Destino",
                                   "Código da Loggi",
                                   "Código do Vendedor do produto",
                                   "Tipo do produto"])

# Começando a indexação com 1
df1.index += 1

# os índices representam o número do pacote
df1
```

```python
# Precisa-se aplicar filtros para que as restrições sejam cumpridas
```

```python
# 1) A Loggi não envia produtos que não sejam dos tipos acima informados.

# Lista com as strings aceitas para Tipo de Produto
valores_aceitos = ["000","111","333","555","888"]

# Selecionando apenas os valores que estão na lista de valores aceitos
df = df1[df1["Tipo do produto"].isin(valores_aceitos)]
```

```python
# 2) Não é possível despachar pacotes contendo jóias tendo como região de origem o Centro-oeste;

# Retirando os códigos de Região não aceitos


# Lista com as strings aceitas para Região
valores_aceitos = ["000","111","333","555","888"]

# Selecionando apenas os valores que estão na lista de valores aceitos
df = df[df["Região de Origem"].isin(valores_aceitos)]
df = df[df["Região de Destino"].isin(valores_aceitos)]
```

```python
# Substituindo valores do código pelos nomes das regiões usando um dicionário

# Região

regiao = {"111" : 'Centro-oeste', "333" : 'Nordeste', "555" : 'Norte', "888" : 'Sudeste', "000" :"Sul"}

df["Região de Origem"].replace(regiao, inplace=True)

df["Região de Destino"].replace(regiao, inplace=True)


# Substituindo valores do código pelos nomes dos produtos usando um dicionário

# Tipo de Produto

produto = {"000" : 'Jóias', "111" : 'Livros', "333" : 'Eletrônicos', "555" : 'Bebidas', "888" :"Brinquedos"}

df["Tipo do produto"].replace(produto, inplace = True)
```

```python
# Removendo linhas onde Região é "Centro-oeste" e onde o produto é "Jóias"

df = df.loc[~((df['Região de Origem'] == "Centro-oeste") & (df['Tipo do produto'] == "Jóias")),:]
```

```python
# 3) O vendedor 584 está com seu CNPJ inativo, códigos de barra relacionados a este vendedor devem ser considerados inválidos.

# Lista com as strings rejeitadas para Código do Vendedor do produto
valores_rejeitados = ["584"]

# Excluindo apenas os valores que estão na lista de valores rejeitados
df = df[~df["Código do Vendedor do produto"].isin(valores_rejeitados)]
```
